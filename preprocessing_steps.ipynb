```python
import pandas as pd

train = pd.read_csv("sales_data_training.csv")
test = pd.read_csv("sales_data_test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
#scaling

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
scaled_train = scaler.fit_transform(train)
scaled_test = scaler.transform(test)
```

```python
scaled_train_df = pd.DataFrame(data=scaled_train,columns = train.columns)
scaled_train_df.head()
```

```python
scaled_test_df = pd.DataFrame(data=scaled_test,columns = test.columns)
scaled_test_df.head()
```

```python
# Print out the adjustment that the scaler applied to the total_earnings column of data
print("Note: total_earnings values were scaled by multiplying by {:.10f} and adding {:.6f}".format(scaler.scale_[8], scaler.min_[8]))
```

```python
#save to csv files
scaled_train_df.to_csv("sales_data_train_scaled.csv",index=False)
scaled_test_df.to_csv("sales_data_test_scaled.csv",index=False)
```
